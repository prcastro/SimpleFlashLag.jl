```julia
using PyPlot, SimpleFlashLag
```

```julia
moving, flash = flashlag_sim();
len, layers, _ = size(moving)
for t in 1:6
    #plot each layer for each time t
    for l in 1:layers
        subplot(6, 6, t + (l-1)*12)
        ylim(-0.5, 3)
        bar(1:len, moving[:,l, 2t], color="black")
        subplot(6, 6, t + (l-1)*12 + 6)
        ylim(-0.5, 3)
        bar(1:len, flash[:,l, 2t], color="black")
    end
end
```
